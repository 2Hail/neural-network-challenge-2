## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df_attr = attrition_df[['Attrition']]
y_df_attr.head()

y_df_dept = attrition_df[['Department']]
y_df_dept.head()
```

```python
attrition_df
```

```python
# Create a list of at least 10 column names to use as X data
X_df = attrition_df[[
    "Education", "Age", "DistanceFromHome",
    "JobSatisfaction","StockOptionLevel", "OverTime", 
    "WorkLifeBalance", "YearsAtCompany", "NumCompaniesWorked"
]]

# Create X_df using your selected columns
# Show the data types for X_df
X_df
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train_attr, y_test_attr, y_train_dept, y_test_dept = train_test_split(X_df, y_df_attr, y_df_dept)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train["OverTime"] = LabelEncoder().fit_transform(X_train["OverTime"])
X_test["OverTime"] = LabelEncoder().fit_transform(X_test["OverTime"])
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_Scaled = scaler.fit_transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
dept_encoder = OneHotEncoder(sparse=False)
# Fit the encoder to the training data

# Create two new variables by applying the encoder
# to the training and testing data

dept_encoder.fit(np.array(y_train_dept['Department']).reshape(-1, 1))
y_department_train = dept_encoder.transform(np.array(y_train_dept['Department']).reshape(-1, 1))
y_department_test = dept_encoder.transform(np.array(y_test_dept['Department']).reshape(-1, 1))

y_department_train
```

```python
# Create a OneHotEncoder for the Attrition column
attr_encoder = OneHotEncoder(sparse=False)
# Fit the encoder to the training data


# Create two new variables by applying the encoder
# to the training and testing data
attr_encoder.fit(np.array(y_df_attr['Attrition']).reshape(-1, 1))

attr_train_encoded = attr_encoder.transform(np.array(y_train_attr['Attrition']).reshape(-1, 1))
attr_test_encoded = attr_encoder.transform(np.array(y_test_attr['Attrition']).reshape(-1, 1))

attr_train_encoded
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_cols = X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input([num_cols,], name="input_layer")

#hidden1 = layers.Dense(units=64, activation='relu', name="hidden1")(input_layer)


#output1 = layers.Dense(units=2,  activation='softmax', name="output1")(hidden1)
# Create at least two shared layers

shared_layer1 = layers.Dense(64, activation='relu', name="shared1")(input_layer)
shared_layer2 = layers.Dense(128, activation='relu', name="shared2")(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
dept_hidden_layer = layers.Dense(32, activation='relu', name="dept_hidden_layer")(shared_layer2)
# Create the hidden layer
dept_output_layer = layers.Dense(3, activation='softmax', name='dept_output_layer')(dept_hidden_layer)

# Create the output layer
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_hidden = layers.Dense(units=32, activation='relu', name='attr_hidden')(shared_layer2)

# Create the output layer
attr_output = layers.Dense(units=2, activation='sigmoid', name='attr_output')(attr_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output_layer, attr_output])

# Compile the model
model.compile(optimizer='adam',
              loss={"attr_output": "binary_crossentropy",
                    "dept_output_layer": "categorical_crossentropy"},
            metrics={"attr_output": "accuracy", 
                    "dept_output_layer": "accuracy"})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {
        "attr_output": attr_train_encoded,
        "dept_output_layer": y_department_train
    },
    epochs=50)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(
    X_test_Scaled,
    {
        "attr_output": attr_test_encoded,
        "dept_output_layer": y_department_test
    })

results
```

```python
# Print the accuracy for both department and attrition
print(f"Dept Pred Acc: {results[2]}")
print(f"Attr Pred Acc: {results[1]}")
```
